```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
cfb = pd.read_csv("CFBeattendance.csv",encoding='unicode_escape')
```

```python
cfb.head()
```

```python
cfb.info()
```

```python
cfb.describe()
```

```python
arr = []
win_team = []
win_rate = []
rank = []
opp_rank = []
tv = []
home_team_win = []
for i in range(len(cfb.values)):
    if 'W' in cfb['Result'][i]:
        arr.append('W')
        home_team_win.append(1)
    else:
        arr.append('L')
        home_team_win.append(0)
    if arr[i] == 'W':
        win_team.append(cfb['Team'][i])
    else:
        win_team.append(cfb['Opponent'][i])
    if cfb['Current Losses'][i] + cfb['Current Wins'][i] != 0:
        win_rate.append((cfb['Current Wins'][i])/(cfb['Current Losses'][i] + cfb['Current Wins'][i]))
    else:
        win_rate.append(0.5)
    if cfb['Rank'][i] == 'NR':
        rank.append(False)
    else:
        rank.append(True)
    if cfb['Opponent_Rank'][i] == 'NR':
        opp_rank.append(False)
    else:
        opp_rank.append(True) 
    if 'Not on TV' in cfb['TV'][i]:
        tv.append(False)
    else:
        tv.append(True)
    
cfb['W/L'] = arr
cfb['win_team'] = win_team
cfb['win_rate'] = win_rate
cfb['Rank'] = rank
cfb['Opponent_Rank'] = opp_rank
cfb['TV'] = tv
cfb['home_team_win'] = home_team_win
```

```python
cfb
```

```python
cols = ['Day','Month','Year','Conference','TMAX','TMIN','SNWD','SNOW','PRCP','Time']
cfb.drop(cols,axis = 1,inplace = True)
```

```python
cfb
```

```python
cfb['Rank'] = cfb['Rank'].astype("category").cat.codes
cfb['TV'] = cfb['TV'].astype("category").cat.codes
cfb['New Coach'] = cfb['New Coach'].astype("category").cat.codes
cfb['Tailgating'] = cfb['Tailgating'].astype("category").cat.codes
cfb['Opponent_Rank'] = cfb['Opponent_Rank'].astype("category").cat.codes
```

```python
cfb
```

```python
cols = ['Date','Team','Opponent','Site','Result','Current Wins','Current Losses','win_team','W/L','Attendance','Stadium Capacity','Tailgating']
train = cfb.drop(cols,axis = 1)
```

```python
train
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train.drop('home_team_win',axis=1), 
                                                    train['home_team_win'], test_size=0.1, 
                                                    random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=55)
```

```python
knn.fit(X_train,y_train)
```

```python
pred = knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,pred))
```
